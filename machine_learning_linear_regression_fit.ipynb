```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(100, size=200)
x2 = x
y2 = x**2 + 10
y2
```

```python
plt.scatter(x2,y2)
plt.show()
```

```python
#to define the cost function: 
def compute_cost(w,b,x2,y2):
    total_cost = 0
    M = len(x2)
    for i in range(M):
        x=x2[i]
        y=y2[i]
        total_cost += (y-w*x-b)**2
    return total_cost/M
```

```python
#to define the average function: 
def average(data):
    sum=0
    num = len(data)
    for i in range(num):
        sum+=data[i]
    return sum/num

def fit(x2,y2):
    M=len(x2)
    x_bar = average(x2)
    # print(x_bar)
    sum_yx = 0
    sum_xsquare = 0
    sum_yw=0

    #here is to calculate w: 
    for i in range(len(x2)):
        x=x2[i]
        y=y2[i]
        sum_yx += y*(x-x_bar)
        sum_xsquare += x**2
    w = sum_yx/(sum_xsquare - M*(x_bar**2))


    #here is to calculate b:
    for i in range(len(x2)):
        x=x2[i]
        y=y2[i]
        sum_yw += (y-w*x)
    b = sum_yw/M

    return w,b
    
```

```python
#test 
w,b = fit(x2,y2)
cost = compute_cost(w,b,x2,y2)
print(f"w:  {w}, b:  {b}, cost : {cost}")
```
