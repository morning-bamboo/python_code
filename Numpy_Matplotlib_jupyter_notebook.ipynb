```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(8,5))
plt.subplot(2,3,2) #2rows,3columns, and current location is 2
plt.plot([0,10],[0,10])
plt.title("2,3,2")

plt.subplot(2,3,4) #2rows,3columns, and current location is 4
plt.plot([0,10],[0,10])
plt.title("2,3,4")

plt.subplot(2,3,5) #2rows,3columns, and current location is 5
plt.plot([0,10],[0,10])
plt.title("2,3,5")

plt.tight_layout()
plt.show()
```

```python
fig,axe = plt.subplots(nrows=3,ncols=3, figsize = (6,6))
axe[1][2].plot([0,10],[0,10], color="red")
axe[1][2].set_title("row2+col3")

plt.tight_layout()

plt.show()
```

```python
x = np.linspace(-3,3,100)
y = np.sin(x)
```

```python
fig,axe1 = plt.subplots()
axe1.plot(x,y,"orange")
axe1.set_xlabel("x")
axe1.set_ylabel("y")
axe1.set_title("Big-Img 1") #to draw a lage image 

plt.show()
```

```python
fig,axe1 = plt.subplots()
axe1.plot(x,y,"orange")
axe1.set_xlabel("x")
axe1.set_ylabel("y")
axe1.set_title("Big-Img 1") #to draw a lage image 

left,bottom,width,height = 0.25,0.6,0.2,0.2
axe2 = fig.add_axes([left,bottom,width,height])
y2=-x**2
axe2.plot(x,y2,"blue")
axe2.set_xlabel("x")
axe2.set_ylabel("y2")
axe2.set_title("Small-Img 1") #to draw a small image 

plt.axes([0.6,0.25,0.2,0.2])
plt.plot(x,x**2,"red")
plt.xlabel("x")
plt.ylabel("y3")
plt.title("Small-Img 2") #to draw a small image 

plt.show()
```

```python
fig,axe = plt.subplots()
label = ["Jan","Feb","Mar","Apr","May","Jun"]

index = np.arange(len(label))
values1 = [100,150,300,200,600,400]
values2 = [150,240,280,220,400,550]
axe.bar(index, values1,width=0.5,label = "values1")

axe.bar(index, values2,width=0.5, bottom=values1, label = "values2")

axe.legend()

plt.show()
```

```python
fig,axe = plt.subplots(figsize = (8,5))

x = np.arange(0, 5, 0.01)
y = np.sin(x * np.pi)
axe.fill_between(x,y,where = (y>0),facecolor = 'red',alpha = 0.7)
axe.fill_between(x,y,where = (y<0),facecolor = 'blue',alpha = 0.6)

plt.show()
```

```python
fig,axe = plt.subplots(figsize = (8,5))
label = ["Jan","Feb","Mar","Apr","May","Jun"]
values = [100,150,300,200,600,400]
explode = [0.3,0,0,0,0.1,0]
axe.pie(values,labels=label,explode = explode, shadow=True, autopct = "%1.2f%%")

plt.show()
```

```python
fig,axe = plt.subplots(figsize = (8,5))
label = ["Jan","Feb","Mar","Apr","May","Jun"]
colors = ["orange","purple","yellow","lightgreen","blue","red"]
values1 = [100,150,300,200,600,400]
values2 = [150,240,280,220,400,550]
explode2 = [0.3,0,0.2,0,0,0]
axe.pie(values1, radius = 1, wedgeprops=dict(width=0.45), colors=colors,labels=label, labeldistance = 0.3, autopct = "%1.2f%%", pctdistance = 0.8) 
axe.pie(values2, radius = 1.5, wedgeprops=dict(width=0.45), colors=colors,labels=label, labeldistance = 1.1, explode = explode2, shadow = False, autopct = "%1.2f%%",pctdistance = 0.8) #larger radius -> outer pie

plt.show()
```

```python
# Create a figure and a single set of axes
fig, ax = plt.subplots() # Correct variable name to ax

values = np.random.randint(0, 5, size=(6, 6))

# Plot the heatmap
heatmap = ax.imshow(values)

# Add the colorbar, stealing space from the 'ax' axes
colorbar = fig.colorbar(heatmap, ax=ax)

# Set the title using the correct method
ax.set_title('Matplotlib Heatmap with Color Bar')

# Display the plot
plt.show()
```
